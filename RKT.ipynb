## LOOKUP FORMULAS

### VLOOKUP(IMPORTANT)

> To finding specific data you want from table

>Takes 4 arguments--lookup_value,table_array,col_index_num,lookup_range

>lookup_value must be in leftmost column

>Values should be in ascending order

>If values are not in ascending order then it will skip the defected cell and try to continue with asc order

>See practice data

#### EXACT MATCH

>Returns exact lookup value

>If value is not present returns error

##### SYNTAX

>=VLOOKUP(value,array,col_num,Exact Match)

>text value should be in ""

#### APPROXIMATE MATCH

> For numbers returns the second largest value in search(for more clarity see examples)

>For alphabets returns the second largest alphabet, keep in mind it checks for every single alphabet

> If it is going out of column  then it will return the last value from column

##### CASE 1:(LOOKING ONLY IN 1 COLUMN)

>syntax: VLOOKUP(value,only 1 column,1,APPROXIMATE MATCH)
    
>As we are only looking in 1 column, col_num will be 1

>if criteria is not satisfied column's last value will be displayed

>syntax--VLOOKUP("alphabet",array,col_num,APPROXIMATE MATCH)
